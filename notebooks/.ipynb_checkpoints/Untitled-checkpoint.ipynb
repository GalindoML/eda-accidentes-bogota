## Limpieza de datos
Importamos pandas y leemos nuestra data inicial

```python
import pandas as pd

pd.set_option('display.max_rows', 10)   
pd.set_option('display.max_columns', 10)

raw_data = pd.ExcelFile("../data/raw/datos-siniestros.xlsx")
```

Extraemos la hoja que vamos a analizar y echamos un vistazo a como esta organizada la informacion

```python
df = pd.read_excel(raw_data, sheet_name = 'Base Consolidada')
pd.DataFrame(df)
```

```python
df = df.dropna(subset='No. IPAT')
df = df.reset_index(drop=True)
pd.DataFrame(df)
```
